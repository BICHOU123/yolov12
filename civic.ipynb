```python
import pandas as pd
from enum import Enum

```

```python
class NodeType(Enum):
    MOLECULAR_PROFILE = "molecular_profile"
    DISEASE = "disease"
    THERAPY = "therapy"
    CLINICAL_SIGNIFICANCE = "clinical_significance"
    EVIDENCE_LEVEL = "evidence_level"
```

```python
class Node:
    def __init__(self, name: str, node_type: NodeType):
        self.name = name.strip()
        self.type = node_type

    def __repr__(self):
        return f"{self.type.name}({self.name})"
```

```python
node_store = {}

def get_or_create_node(name: str, node_type: NodeType):
    key = (name.strip(), node_type)
    if key not in node_store:
        node_store[key] = Node(name, node_type)
    return node_store[key]
```

```python
df = pd.read_csv("/content/civic_hg_rag_clean (1).csv",sep=",")
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace("\ufeff", "", regex=False)
)
```

```python
print(df.columns.tolist())
```

```python
for _, row in df.iterrows():

    get_or_create_node(
        row["molecular_profile"],
        NodeType.MOLECULAR_PROFILE
    )

    get_or_create_node(
        row["disease"],
        NodeType.DISEASE
    )

    get_or_create_node(
        row["therapies"],
        NodeType.THERAPY
    )

    get_or_create_node(
        row["significance"],
        NodeType.CLINICAL_SIGNIFICANCE
    )

    get_or_create_node(
        row["evidence_level"],
        NodeType.EVIDENCE_LEVEL
    )
```
